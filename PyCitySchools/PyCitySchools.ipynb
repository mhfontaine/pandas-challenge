```python
import pandas as pd
```

```python
file_1 = "schools_complete.csv"
file_2 = "students_complete.csv"
```

```python
# Read student data file
studentdata_df = pd.read_csv(file_2)
#studentdata_df.head()

# Read school data file
schoolsdata_df = pd.read_csv(file_1)

# merged data frames
schoolsdata_df.rename(columns = {'name': 'school'}, inplace = True)

merged_df = studentdata_df.merge(schoolsdata_df, how = 'left', on = 'school_name')

file_3 = pd.merge(studentdata_df, schoolsdata_df, how = 'left', on = ['school_name', 'school_name'])
```

# District Summary

```python
# District Summaries

# Total Schools
total_schools = len(file_3["school_name"].unique())

# Total Students
total_students = len(file_3["Student ID"].unique())

# Total Budget
sum_budget = schoolsdata_df["budget"].sum()
sumbudget = "${:,}".format(sum_budget)

# Average Math Score
average_math = file_3["math_score"].mean()

# Average Reading Score
average_reading = file_3["reading_score"].mean()

# Students passing math
passing_math_score = file_3[(file_3["math_score"]>=70)].count()["Student ID"]

# Students passing math %
passing_math_percent = (passing_math_score / total_students) * 100

# Students reading %
passing_reading_percent = (passing_reading_count / total_students) * 100

# Students passing math and reading %
overall_count = file_3[ (file_3["math_score"]>=70)  & (file_3["reading_score"]>=70)   ].count()["student_name"]

# Overall passing rate
total_percent = (overall_count / total_students) * 100

# District Summary Table
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students ": [total_students],
    "Total Budget": [sumbudget],
    "Average Math Scores" : average_math,
    "Average Reading Scores" : average_reading,
    "% Passing Math" : passing_math_percent,
    "% Passing Reading" : passing_reading_percent,
    "% Overall Passing" : total_percent
    })
district_summary
```

# School Summary

```python
# Top Performing Schools
by_school = file_3.set_index('school_name').groupby(['school_name'])

# School Types
sch_types = schoolsdata_df.set_index('school_name')['type']

# Total Students Per School
stu_per_sch = by_school['Student ID'].count()

# School Budget
sch_budget = schoolsdata_df.set_index('school_name')['budget']

# Budget Per Student
stu_budget = schoolsdata_df.set_index('school_name')['budget']/schoolsdata_df.set_index('school_name')['size']

# Average Math Score
avg_math = by_school['math_score'].mean()

# Average Reading Score
avg_read = by_school['reading_score'].mean()

passmath = file_3[file_3['math_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_sch*100

passreading = file_3[file_3['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_sch*100


# Overall Passing Rate
overallpass = (passmath + passreading)/2


school_summary = pd.DataFrame({
    "School Type": sch_types,
    "Total Students": stu_per_sch,
    "Per Student Budget": stu_budget,
    "Total School Budget": sch_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "Percentage Passing Math" : passmath,
    "Percentage Passing Reading" : passreading,
    "Percentage Overall Passing" : overallpass
})
school_summary.head()
```

# Scores by School Spending

```python
# Spending Bins

bins = [0, 584.999, 614.999, 644.999, 999999]
group_name = ['< $585', "$585 - 614", "$615 - 644", "> $644"]
merged_df['spending_bins'] = pd.cut(merged_df['budget']/merged_df['size'], bins, labels = group_name)

# Spending Groups
by_spending = merged_df.groupby('spending_bins')

# Calculations
avg_math = by_spending['math_score'].mean()
avg_read = by_spending['reading_score'].mean()
pass_math = merged_df[merged_df['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
pass_read = merged_df[merged_df['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
overall = merged_df[(merged_df['reading_score'] >= 70) & (merged_df['math_score'] >= 70)].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()

            
# DataFrame            
scores_by_spend = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall
            
})
            
# Columns
scores_by_spend = scores_by_spend[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "Overall Passing Rate"
]]

scores_by_spend.index.name = "Per Student Budget"
scores_by_spend = scores_by_spend.reindex(group_name)


scores_by_spend.style.format({'Average Math Score': '{:.1f}', 
                              'Average Reading Score': '{:.1f}', 
                              '% Passing Math': '{:.1%}', 
                              '% Passing Reading':'{:.1%}', 
                              'Overall Passing Rate': '{:.1%}'})
```

# Scores by School Size

```python
# Bins by S,M,L

bins = [0, 999, 1999, 99999999999]
group_name = ["Small (<1000)", "Medium (1000-2000)" , "Large (>2000)"]
merged_df['size_bins'] = pd.cut(merged_df['size'], bins, labels = group_name)

# Spending Groups
by_size = merged_df.groupby('size_bins')

# Calculations 
avg_math = by_size['math_score'].mean()
avg_read = by_size['math_score'].mean()
pass_math = merged_df[merged_df['math_score'] >= 70].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()
pass_read = merged_df[merged_df['reading_score'] >= 70].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()
overall = merged_df[(merged_df['reading_score'] >= 70) & (merged_df['math_score'] >= 70)].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()

            
# DataFrame           
scores_by_size = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall
            
})
            
# Columns
scores_by_size = scores_by_size[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "Overall Passing Rate"
]]

scores_by_size.index.name = "Total Students"
scores_by_size = scores_by_size.reindex(group_name)

#
scores_by_size.style.format({'Average Math Score': '{:.1f}', 
                              'Average Reading Score': '{:.1f}', 
                              '% Passing Math': '{:.1%}', 
                              '% Passing Reading':'{:.1%}', 
                              'Overall Passing Rate': '{:.1%}'})
```
